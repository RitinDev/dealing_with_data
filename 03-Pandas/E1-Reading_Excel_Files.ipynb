## Reading Excel files

Pandas make it trivially easy to read the contents of Excel files. For example, I stored the restaurant inspection dataset as an excel file. Let's grab it and get it stored locally:

```python
!curl 'https://data.cityofnewyork.us/api/views/43nn-pn8j/files/ec33d2c8-81f5-499a-a238-0213a38239cd?download=true&filename=RestaurantInspectionDataDictionary_09242018.xlsx' -o data_dictionary.xlsx
```

```python
import pandas as pd

restaurants_data_dictionary = pd.read_excel ("data_dictionary.xlsx", sheet_name=1, header=1);
```

```python
restaurants_data_dictionary.columns = ['Column_Name', 'Description', 'Code_Definitions', 'Notes']
```
